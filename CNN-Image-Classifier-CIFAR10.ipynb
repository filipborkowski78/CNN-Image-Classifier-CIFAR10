```python
# -*- coding: utf-8 -*-
# =========================================================
# --- 0 : Montowanie Google Drive ---
# =========================================================
from google.colab import drive
import os

# Montowanie dysku Google
drive.mount('/content/drive')

# Ścieżka do folderu roboczego na Drive
base_export_dir = '/content/drive/MyDrive/PRO1'
os.makedirs(base_export_dir, exist_ok=True)

print(f"[OK] Google Drive podłączony. Pliki będą zapisywane w:\n{base_export_dir}")

# =========================================================
# --- 0. Import bibliotek ---
# =========================================================
import numpy as np
import matplotlib.pyplot as plt
import datetime

from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import (
    Input,
    Conv2D,
    BatchNormalization,
    Activation,
    MaxPooling2D,
    GlobalAveragePooling2D,
    Flatten,
    Dense,
    Dropout,
    RandomFlip,
    RandomRotation,
    RandomZoom
)
from keras.optimizers import Adam
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from keras import regularizers

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# =========================================================
# --- 1. Importowanie danych CIFAR10 ---
# =========================================================
# Wczytanie zbioru danych CIFAR10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Wyświetlenie przykładowego obrazu
print("Etykieta przykładowego obrazu:", y_train[0])
plt.imshow(x_train[0])
plt.axis('off')
plt.show()

# =========================================================
# --- 2. Wstępne przetwarzanie danych ---
# =========================================================
# Normalizacja do zakresu [0,1]
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32') / 255.0

# One-hot encoding etykiet (liczby -> wektory [0.0.0.3.0.0])
num_classes = 10
y_train_cat = to_categorical(y_train, num_classes)
y_test_cat  = to_categorical(y_test, num_classes)

# lista nazw klas
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

# =========================================================
# --- 3. Definicja modelu sieci neuronowej ---
# =========================================================
input_shape = (32, 32, 3)  # CIFAR-10 to 32x32 RGB

# weight_decay = siła regularyzacji L2,
# czyli „kara” za zbyt duże wagi – pomaga walczyć z przeuczeniem.
weight_decay = 1e-4

model = Sequential([
    Input(shape=input_shape),

    # Augmentacja danych w modelu
    RandomFlip("horizontal"),
    RandomRotation(0.1),
    RandomZoom(0.1),

    # Blok 1
    Conv2D(32, (3,3), padding='same',
           kernel_regularizer=regularizers.l2(weight_decay)),#wyciąga cechy obrazu
    BatchNormalization(),               # stabilizuje rozkład aktywacji
    Activation('relu'),                 # nieliniowość
    Conv2D(32, (3,3), padding='same',
           kernel_regularizer=regularizers.l2(weight_decay)),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2,2)),                # zmniejsza rozmiar obrazu o połowę
    Dropout(0.25),                      # losowo wyłącza 25% neuronów – mniejsze przeuczenie

    # Blok 2
    Conv2D(64, (3,3), padding='same',
           kernel_regularizer=regularizers.l2(weight_decay)),
    BatchNormalization(),
    Activation('relu'),
    Conv2D(64, (3,3), padding='same',
           kernel_regularizer=regularizers.l2(weight_decay)),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2,2)),
    Dropout(0.25),

    # Blok 3
    Conv2D(128, (3,3), padding='same',
           kernel_regularizer=regularizers.l2(weight_decay)),
    BatchNormalization(),
    Activation('relu'),
    Conv2D(128, (3,3), padding='same',
           kernel_regularizer=regularizers.l2(weight_decay)),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D((2,2)),
    Dropout(0.25),

    # Zamiast Flatten używam GlobalAveragePooling2D:
    # każdą mapę cech ściskam do jednej średniej wartości.
    # Dzięki temu mam mniej parametrów i mniejsze ryzyko przeuczenia.
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),                 #żeby się nie przeuczyć
    #10 neuronów po jednym na klase, softmax - wyniki -> rozkład prawdopobonieństwa
    Dense(num_classes, activation='softmax')
])

# =========================================================
# --- 4. Kompilacja modelu ---
# =========================================================
# Optymalizator Adam, funkcja straty categorical_crossentropy, metryka accuracy
learning_rate = 1e-3
optimizer = Adam(learning_rate=learning_rate)

model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Wyświetlenie podsumowania modelu
model.summary()

# =========================================================
# --- 5. Trenowanie modelu ---
# =========================================================

es = EarlyStopping(
    monitor='val_loss',     # obserwujemy val_loss
    patience=10,            # ile epok bez poprawy zanim zatrzymamy
    min_delta=1e-4,         # minimalna wymagana zmiana, by uznać, że jest „poprawa”
    mode='min',             #  im MNIEJSZY val_loss, tym lepiej
    restore_best_weights=True  # po zatrzymaniu przywróć najlepsze wagi
)

rlp = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,             # zmniejszam LR o połowę
    patience=3,             # czekam 3 epoki bez poprawy
    min_delta=1e-4,
    min_lr=1e-5,            # nie schodzę poniżej 1e-5
    mode='min'
)

checkpoint_path = os.path.join(base_export_dir, "best_model_val_loss.keras")
mcp = ModelCheckpoint(
    filepath=checkpoint_path,
    monitor='val_loss',
    save_best_only=True,
    mode='min'
)

batch_size = 128            # ile obrazków naraz w jednej „porcji”
epochs = 80                 # maksymalna liczba epok EA i tak wcześniej zatrzyma

history = model.fit(
    x_train,
    y_train_cat,
    batch_size=batch_size,
    epochs=epochs,
    # Używam testu jako validation_data (zgodnie z template),
    # dzięki temu mam od razu „val_loss na zbiorze testowym”.
    validation_data=(x_test, y_test_cat),
    callbacks=[es, rlp, mcp],
    verbose=1
)

# Timestamp – żeby pliki miały unikalne nazwy (data+godzina)
ts = datetime.datetime.now().strftime("_%Y%m%d_%H%M")

# =========================================================
# --- 6. Ewaluacja modelu ---
# =========================================================
# Tutaj liczę finalne wyniki modelu na zbiorze testowym.
loss, accuracy = model.evaluate(x_test, y_test_cat, verbose=0)
print('Dokładność na zbiorze testowym:', f"{accuracy:.4f}")
print('Strata na zbiorze testowym:', f"{loss:.4f}")

# =========================================================
# --- 7. Wizualizacja przebiegu treningu (loss & accuracy) ---
# =========================================================
fig_loss_acc = plt.figure(figsize=(10,4))
# Wykres loss
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='loss (train)')
plt.plot(history.history['val_loss'], label='loss (val)')
plt.xlabel('Epoka')
plt.ylabel('categorical_crossentropy')
plt.title(f'Strata na zbiorze testowym to {loss:.4f}')
plt.legend()
# Wykres accuracy
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='accuracy (train)')
plt.plot(history.history['val_accuracy'], label='accuracy (val)')
plt.xlabel('Epoka')
plt.ylabel('Accuracy')
plt.title(f"Dokładność na zbiorze testowym to {accuracy:.4f}")
plt.legend()
plt.tight_layout()

# Zapis wykresu
loss_acc_path = os.path.join(base_export_dir, f'training_loss_accuracy{ts}.png')
fig_loss_acc.savefig(loss_acc_path, dpi=300)
print("Zapisano wykres loss/accuracy ->", loss_acc_path)

plt.show()

# =========================================================
# --- 8. Wizualizacja błędnych klasyfikacji i macierz pomyłek ---
# =========================================================
# Predykcje
pred_probs = model.predict(x_test, batch_size=256, verbose=1)
pred_labels = np.argmax(pred_probs, axis=1)
true_labels = np.argmax(y_test_cat, axis=1)

# Indeksy przykładów, w których model się pomylił
incorrect_indices = np.nonzero(pred_labels != true_labels)[0]

# Wyświetlenie kilku błędnych przykładów
n_show = min(3, len(incorrect_indices))
for i in range(n_show):
    idx = incorrect_indices[i]
    plt.figure(figsize=(3,3))
    plt.imshow(x_test[idx])
    plt.title(f"Prawidłowo: {class_names[true_labels[idx]]}  -> Predykcja: {class_names[pred_labels[idx]]}")
    plt.axis('off')
    plt.show()

#MAcierz pomyłek
cm = confusion_matrix(true_labels, pred_labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
fig_cm, ax = plt.subplots(figsize=(8,8))
disp.plot(ax=ax, cmap='Blues', colorbar=False)
plt.title('Macierz pomyłek')

# Zapis wykresu PNG
cm_path = os.path.join(base_export_dir, f'confusion_matrix{ts}.png')
fig_cm.savefig(cm_path, dpi=300)
print("Zapisano macierz pomyłek ->", cm_path)
plt.show()

# =========================================================
# --- 9. Zapis modelu ---
# =========================================================
# Zapis ostatniej wersji modelu (po EarlyStopping)
model_path = os.path.join(base_export_dir, f"my_model{ts}.keras")
model.save(model_path)
print("Model zapisany jako", model_path)
print("Najlepszy model wg val_loss zapisano jako:", checkpoint_path)
```
